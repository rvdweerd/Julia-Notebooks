Suppose we want to solve f(x)=0 with numerical iteration for x in R1 and f(x) a polynomial, say f(x)=x^2-9.

Newton's method:
x_k+1 = x_k - J(x_k)^-1 * f(x_k) 
With J = Jacobian J_j_k = df_j/dx_k (partial derivatives)

In our example, J = 2x

```julia
using LinearAlgebra
using Plots
```

```julia
function F(x)
    return x*x-9;
end
```

```julia
function J(x)
    return 2*x;
end
```

```julia
x_ = 0.5;
y=[];
insert!(y,1,x_);
for i=2:5
    x__ = x_ - 1/J(x_)*F(x_)
    insert!(y,i,x__)
    x_=x__
end
```

```julia
y
```

```julia
plot(y)
```
