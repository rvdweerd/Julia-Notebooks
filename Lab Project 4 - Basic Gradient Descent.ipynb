```julia
using Plots
```

```julia
x=-3:0.05:5
y=-3:0.05:5
b=.2;
```

```julia
f(x,y) = 0.5*(x*x+b*y*y)
```

```julia
X=repeat(reshape(x,1,:),length(y),1);
Y=repeat(y,1,length(x));
Z=map(f,X,Y);
```

```julia
p1=contour(x,y,Z,levels=[0.01,0.05,0.1,0.25,0.5,1,2,f(3,4)])
```

```julia
x0=[3; 4]
s=1.5
A=[1-s 0 ; 0 1-s*b]
plot!([x0[1]],[x0[2]],seriestype=:scatter);
```

```julia
x_=[]
y_=[]
for i in 1:10
    global A,x0,p1,x_,y_
    xi=A*x0
    insert!(x_,i,xi[1])
    insert!(y_,i,xi[2])
    println(xi)
    plot!([xi[1]],[xi[2]],seriestype=:scatter,smooth=true)  
    
    x0=xi
end
#println(x_)
#println(y_)
#plot(x_,y_,seriestype=:line)
```
