```julia
using LinearAlgebra
using Plots
```

```julia
x = [ [1 5 20]', [1, 2.5, 8] , [1 5.5 9]' , [1 7.7 20]', [1 9 12]', [1  2 25]', [1 4 30]', [1  5.1 24]', [1 6.6 37]']
y = [ 1,1,1,1,1,-1,-1,-1,-1]
w= ones(3)
x1=[];
x2=[];
```

```julia
for i=1:9
    insert!(x1,length(x1)+1,x[i][2])
    insert!(x2,length(x2)+1,x[i][3])
end
```

```julia
plot(x1[1:5],x2[1:5],seriestype=:scatter)
plot!(x1[6:9],x2[6:9],seriestype=:scatter)
```

```julia
function IndexOfWrongPrediction(w,x,y)
    for i=1:length(x)
        v = w'*x[i]
        if (v[1]/abs(v[1])) != y[i]
            return i
        end
    end
    return -1
end
```

```julia
itCount = 0;
i=IndexOfWrongPrediction(w,x,y)
while i>0
    global w,y,x,i,itCount 
    w =  w + y[i]*x[i]
    i=IndexOfWrongPrediction(w,x,y)
    itCount=itCount+1;
end
```

```julia
println(itCount)
```

```julia
alpha = -w[2]/w[3]
beta = -w[1]/w[3]
t=0:1:10
plot!(t,t->alpha*t+beta)
```
